```r
#Mode doesn't have an in-built function hence we have to build one
getmode <- function(v) {
   uniqv <- unique(v)
   uniqv[which.max(tabulate(match(v, uniqv)))]
}
```

```r
eruptions <- faithful$eruptions
```

```r
waiting <- faithful$waiting
```

# **Bivariate Analysis**

```r
cov(eruptions, waiting)
#Indicates a positive linear rship btwn the 2 variables
```

```r
cor(eruptions, waiting)
#Positive linearly related
```

```r
#Creating A Scatter Plot
plot(eruptions, waiting, xlab="Eruption duration", ylab="Time waited")
#Positive linear rship btwn eruptions and waiting
```

```r
cor(faithful, method = c("pearson", "kendall", "spearman"))
```

```r
faithful.cor = cor(faithful)
```

```r
faithful.cor = cor(faithful, method = c("spearman"))
```

```r
install.packages("Hmisc")
library("Hmisc")
```

```r
faithful.rcorr = rcorr(as.matrix(faithful))
faithful.rcorr
```

```r
faithful.coeff = faithful.rcorr$r
faithful.p = faithful.rcorr$P
```
